```python
# Encoding
CentralAir = House_data['CentralAir'].unique()
encoding_dict = {value: i for i, value in enumerate(CentralAir)}
House_data['CentralAir'] = House_data['CentralAir'].apply(lambda x: encoding_dict[x])

LandSlope = House_data['LandSlope'].unique()
encoding_dict = {value: i for i, value in enumerate(LandSlope)}
House_data['LandSlope'] = House_data['LandSlope'].apply(lambda x: encoding_dict[x])

Utilities = House_data['Utilities'].unique()
encoding_dict = {value: i for i, value in enumerate(Utilities)}
House_data['Utilities'] = House_data['Utilities'].apply(lambda x: encoding_dict[x])

LotShape = House_data['LotShape'].unique()
encoding_dict = {value: i for i, value in enumerate(LotShape)}
House_data['LotShape'] = House_data['LotShape'].apply(lambda x: encoding_dict[x])

Street = House_data['Street'].unique()
encoding_dict = {value: i for i, value in enumerate(Street)}
House_data['Street'] = House_data['Street'].apply(lambda x: encoding_dict[x])

Neighborhood = House_data['Neighborhood'].unique()
encoding_dict = {value: i for i, value in enumerate(Neighborhood)}
House_data['Neighborhood'] = House_data['Neighborhood'].apply(lambda x: encoding_dict[x])

HouseStyle = House_data['HouseStyle'].unique()
encoding_dict = {value: i for i, value in enumerate(HouseStyle)}
House_data['HouseStyle'] = House_data['HouseStyle'].apply(lambda x: encoding_dict[x])

RoofMatl = House_data['RoofMatl'].unique()
encoding_dict = {value: i for i, value in enumerate(RoofMatl)}
House_data['RoofMatl'] = House_data['RoofMatl'].apply(lambda x: encoding_dict[x])

Exterior2nd = House_data['Exterior2nd'].unique()
encoding_dict = {value: i for i, value in enumerate(Exterior2nd)}
House_data['Exterior2nd'] = House_data['Exterior2nd'].apply(lambda x: encoding_dict[x])

ExterCond = House_data['ExterCond'].unique()
encoding_dict = {value: i for i, value in enumerate(ExterCond)}
House_data['ExterCond'] = House_data['ExterCond'].apply(lambda x: encoding_dict[x])

Foundation = House_data['Foundation'].unique()
encoding_dict = {value: i for i, value in enumerate(Foundation)}
House_data['Foundation'] = House_data['Foundation'].apply(lambda x: encoding_dict[x])

HeatingQC = House_data['HeatingQC'].unique()
encoding_dict = {value: i for i, value in enumerate(HeatingQC)}
House_data['HeatingQC'] = House_data['HeatingQC'].apply(lambda x: encoding_dict[x])

KitchenQual = House_data['KitchenQual'].unique()
encoding_dict = {value: i for i, value in enumerate(KitchenQual)}
House_data['KitchenQual'] = House_data['KitchenQual'].apply(lambda x: encoding_dict[x])

Functional = House_data['Functional'].unique()
encoding_dict = {value: i for i, value in enumerate(Functional)}
House_data['Functional'] = House_data['Functional'].apply(lambda x: encoding_dict[x])

GarageType = House_data['GarageType'].unique()
encoding_dict = {value: i for i, value in enumerate(GarageType)}
House_data['GarageType'] = House_data['GarageType'].apply(lambda x: encoding_dict[x])

GarageCond = House_data['GarageCond'].unique()
encoding_dict = {value: i for i, value in enumerate(GarageCond)}
House_data['GarageCond'] = House_data['GarageCond'].apply(lambda x: encoding_dict[x])

PoolQC = House_data['PoolQC'].unique()
encoding_dict = {value: i for i, value in enumerate(PoolQC)}
House_data['PoolQC'] = House_data['PoolQC'].apply(lambda x: encoding_dict[x])

Fence = House_data['Fence'].unique()
encoding_dict = {value: i for i, value in enumerate(Fence)}
House_data['Fence'] = House_data['Fence'].apply(lambda x: encoding_dict[x])

MiscFeature = House_data['MiscFeature'].unique()
encoding_dict = {value: i for i, value in enumerate(MiscFeature)}
House_data['MiscFeature'] = House_data['MiscFeature'].apply(lambda x: encoding_dict[x])

SaleType = House_data['SaleType'].unique()
encoding_dict = {value: i for i, value in enumerate(SaleType)}
House_data['SaleType'] = House_data['SaleType'].apply(lambda x: encoding_dict[x])

SaleCondition = House_data['SaleCondition'].unique()
encoding_dict = {value: i for i, value in enumerate(SaleCondition)}
House_data['SaleCondition'] = House_data['SaleCondition'].apply(lambda x: encoding_dict[x])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()

# Drop unnecessary features
X = House_data.loc[:,['Street', 'LotShape', 'Utilities', 'LandSlope', 'Neighborhood',
       'HouseStyle', 'RoofMatl', 'Exterior2nd', 'ExterCond', 'Foundation',
       'HeatingQC', 'CentralAir', 'KitchenQual', 'Functional', 'GarageType',
       'GarageCond', 'PoolQC', 'Fence', 'MiscFeature', 'SaleType',
       'SaleCondition']]

y = House_data['SalePrice']

# Initialize scaler
X_scale = scale.fit_transform(X)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scale, y, test_size=0.2, random_state=42)
```
